```python

import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Question 1 : plot a line plot between a and b
a=np.arange(40,50)
b=np.arange(50,60)

plt.plot(a,b, 'go-')
plt.title('Line Plot between "a" and "b"')
plt.xlabel('a')
plt.ylabel('b')
plt.show()
```

```python
# Question 2 : Plot a line plot showing the sales trend in company 1 and 2

days = [1,2,3,4,5,6,7] 
sales_1 = [160,150,140,145,175,165,180] 
sales_2 = [70,90,160,150,140,145,175]  

plt.plot(days,sales_1,"g--", label = "sales 1")
plt.plot(days,sales_2,'r-', label = "sales 2")
plt.xlabel('Days')
plt.ylabel('Sales')
plt.title("Sales Graph")
plt.legend()
plt.show()
```

```python
# Question 3 : Create a 3 by 3 subplots:
plt.figure(figsize = (10,10))
x = [1,2,3,4]
y1 = [4,3,2,1]
y2 = [10,20,30,40]
y3 = [40,30,20,10]
y4 = [1,2,1,2]
y5 = [40,70,90,70]

plt.subplot(3,3,1)
plt.plot(x,y1, 'r--')
plt.subplot(3,3,2)
plt.plot(x,y2, 'g*--')
plt.subplot(3,3,3)
plt.plot(x,y3, 'bo-')
plt.subplot(3,3,4)
plt.plot(x,y4, 'orange')
plt.subplot(3,3,5)
plt.plot(x,y5, 'yellow')
plt.show()
```
